## Задача 1:

Дан список из чисел.
Определите их НОК (наименьшее общее кратное) и НОД (наибольший общий делитель).

```python
def gcd_euclid(a, b):
    while b:
        a, b = b, a % b
    return a
     
def calculate_lcm(x, y): 
    if x > y:
        greater = x
    else:
        greater = y
    while(True):
        if((greater % x == 0) and(greater % y == 0)):
            lcm = greater
            break
        greater += 1 
    return lcm


def gcd_multiple_euclid(numbers):
    result = numbers[0]
    for i in range(1, len(numbers)):
        result = gcd_euclid(result, numbers[i])
    return result
     

def gcd_lcm(numbers):
    result = numbers[0]
    for i in range(1, len(numbers)):
        result = calculate_lcm(result, numbers[i])
    return result
     
print("НОД: ", gcd_multiple_euclid([360, 480, 4800])) 
print("НОК: ", gcd_lcm([360, 480, 4800])) 
```

## Задача 2:

Даны n предложений. Определите, сколько из них содержат хотя бы одну цифру.

```python
import re

text_list = ["Мы купили 2 банана", "По реке плывет кирпич", "На дворе осень 2023"]

def num_count(spisok):
    spisok1 = [re.sub(r'[^a-zA-Zа-яА-Я0-9]', '', i) for i in spisok]
    spisok_bool = [i.isalpha() for i in spisok1]
    return [spisok_bool.count(i) for i in set(spisok_bool) if i == False][0]

print("Количество предложений, содержащих хотя бы одну цифру: ", num_count(text_list))
```

Для введенного предложения выведите статистику символ=количество. Регистр букв не учитывается.

```python
text = input("Введите предложение: ")

def statistic(text):
    sym = [char for char in text.lower()]
    return {i: sym.count(i) for i in set(sym)}

statistic(text)
```

## Задача 5:

Используя шифр Цезаря (достаточно только букв русского алфавита, знаки препинания не изменяются), зашифруйте, а затем расшифруйте введенную строку.

```python
alf = "АБВГДЕЁЖЗИЙКЛМНОПРСТУФХЦЧШЩЪЫЬЭЮЯ"

text = input("Введите строку: ")

delta = 7

def split(s):
    
    return [char for char in s]

def shifr(text):
    sh = []
    for i in split(text.upper()):
        for idx2,j in enumerate(split(alf)):
            if i == j:
                idx2 = (idx2+delta)%len(alf)
                sh.append(alf[idx2])
    return sh

def deshifr(sh):
    desh = []
    for i in sh:
        for idx2,j in enumerate(split(alf)):
            if i == j:
                idx2 = (idx2-delta)%len(alf)
                desh.append(alf[idx2])
    return(desh)

            
print("Зашифрованная строка: ", ''.join(shifr(text)))
print("Расшифрованная строка: ", ''.join(deshifr(shifr(text))))
```

## Задача 6:

Напишите функцию, которая принимает неограниченное количество числовых аргументов и возвращает кортеж из двух списков:

* отрицательных значений (отсортирован по убыванию);
* неотрицательных значений (отсортирован по возрастанию).


```python
l = []

while True:
    n = input("Введите число по одному. Когда закончите, нажмите Enter ")
    if n == "":
        break
    else:
        l.append(n)
        
l = [int(i) for i in l]

def kortezh(spisok):
    
    pol = []
    otr = []

    for i in spisok:
        if i >= 0:
            pol.append(i)
        elif i<0:
            otr.append(i)
    return sorted(pol), sorted(otr, reverse = True)  

print("Отсортированные кортежи: ", kortezh(l))
```

## Задача 7:

Строка называется палиндромом, если она пишется одинаково в обоих направлениях. Например, палиндромами в английском языке являются слова «anna», «civic», «level», «hannah». Напишите программу, запрашивающую у пользователя строку и при помощи цикла определяющую, является ли она палиндромом.

```python
string = input("Введите слово: ")

def split(s):
    return [char for char in s]

def palindrom(s):
    if split(s)[::-1] == split(s):
        print(f"Слово {string} - палиндром")
    else:
        print(f"Слово {string} - не палиндром")
        
palindrom(string)
```

1 — равно,\
2 — больше,\
3 — меньше.

Напишите программу, которая с помощью цепочки таких вопросов и ответов угадывает число.
Дополнительно: сделайте так, чтобы можно было гарантированно угадать число за семь попыток.

```python
def bin_search(list):
    
    mid = 1
    left = 1
    right = 100
    
    while left <= right:
        mid = (left + right) // 2
        k = int(input(f"Твое число равно, больше или меньше, чем {mid}? Ответь 1 - равно, 2 - больше, 3 - меньше "))
        if k == 1:
            return mid
        elif k == 2:
            left = mid+1
        elif k == 3:
            right = mid-1
            
    return mid

print(f"Поиск окончен. Твое число {bin_search(range(1,101))}")
```

```python
#Перевод числа из любой системы в десятичную

def to_hex(sys_in, sys_out, n):

    dict = {"A": 10, "B": 11, "C": 12, "D": 13, "E": 14, "F": 15}

    n_list = list(','.join(str(n)[::-1].split()))

    sum=0

    for idx, i in enumerate(n_list):
        for key, value in dict.items():
            if i == key:
                n_list[idx] = n_list[idx].replace(i, str(value))
            else:
                pass
         
    for idx, i in enumerate(n_list):
        sum+=int(i)*sys_in**idx 
    
    return sum
```

```python
#Перевод числа из десятичной системы в любую

def from_hex(sys_in, sys_out, n):

    dict = {"A": 10, "B": 11, "C": 12, "D": 13, "E": 14, "F": 15}

    l = []

    while True:
        l.append(n%sys_out)
        if n//sys_out == 0:
            break
        else:
            n=n//sys_out
        
        
    x = [str(i) for i in l[::-1]]  
    
    for idx, i in enumerate(x):
        for key, value in dict.items():
            if i == str(value):
                x[idx] = x[idx].replace(i, key)
            else:
                pass
            

    if all([i.isdigit() for i in x]):
        return int(''.join(x))
    else:
        return ''.join(x)
```

## Задача 10:

Магическими называются даты, в которых произведение дня и месяца составляет последние две цифры года. Например, 10 июня 1960 года – магическая дата, поскольку 10 ´ 6 = 60. Напишите функцию, определяющую, является ли введенная дата магической. Используйте написанную функцию в главной программе для отображения всех магических дат в XX ве­ке.

```python
import datetime

start_date = datetime.date(1901,1,1)
end_date = datetime.date(2000,12,31)
step = datetime.timedelta(days=1)

dates = []
mag_dates= []

def is_magical_date(date):
    if date.day*date.month == int(str(date.year)[-2:]):
        mag_dates.append(date)
        print("Эта дата - магическая!")
        return mag_dates
    else:
        print("Эта дата - не магическая")

        
while start_date <= end_date:
    dates.append(start_date)
    start_date += step

for date in dates:
    if date.day*date.month == int(str(date.year)[-2:]):
        mag_dates.append(date)

[i.strftime("%Y-%m-%d") for i in mag_dates]
```
